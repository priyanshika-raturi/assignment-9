Question 1: What is the difference between descriptive statistics and inferential
statistics? Explain with examples.


Statistics is a branch of mathematics that deals with the collection, analysis, interpretation, and presentation of data. It helps in understanding large sets of information and drawing useful conclusions. Statistics is mainly divided into **two major types**: *Descriptive Statistics* and *Inferential Statistics*. Both play an important role in research and data analysis, but they serve different purposes.

---

 **Descriptive Statistics**

Descriptive statistics refers to methods used to **describe, summarize, and present** the main features of a dataset in a meaningful way. It gives a simple summary of the collected data without drawing conclusions beyond the data itself. It focuses on the past or present data.

Common tools used in descriptive statistics include:

* **Measures of central tendency**: Mean, Median, Mode
* **Measures of dispersion**: Range, Variance, Standard Deviation
* **Data representation methods**: Tables, Graphs, Histograms, Pie charts

**Example:**
Suppose marks of 100 students in a mathematics exam were collected and the calculated **average score (mean)** is 72 marks. This value simply summarizes the data collected from those exact 100 students. It does not predict or generalize about other students.
Another example: In a survey of 50 customers, **60% are satisfied** with a product. This percentage describes the existing data only.

---

 **Inferential Statistics**

Inferential statistics refers to methods used to **make predictions, decisions, or generalizations** about a larger population based on data obtained from a sample. Since it deals with prediction, there is always some degree of uncertainty. It uses probability concepts and statistical models to draw conclusions.

Common techniques used in inferential statistics include:

* Hypothesis testing
* Confidence intervals
* Regression analysis
* Correlation
* Chi-square test, t-test, ANOVA, etc.

**Example:**
A researcher selects a sample of **30 students** from a school and finds their average height to be **160 cm**. Using inferential statistics, the researcher may **predict that the average height of all 1,000 students in the school** is around 160 cm, with a certain confidence level (e.g., 95% confidence interval).
Another example is predicting election results based on a survey of 2,000 voters from a population of millions.



### **Key Differences Between Descriptive and Inferential Statistics**

| **Descriptive Statistics**              | **Inferential Statistics**                                  |
| --------------------------------------- | ----------------------------------------------------------- |
| Summarizes and describes existing data. | Draws conclusions and makes predictions about a population. |
| Based on complete data.                 | Based on sample data.                                       |
| Deals with present and past.            | Deals with future possibilities.                            |
| No uncertainty or probability involved. | Involves uncertainty and uses probability.                  |
| Example: Average marks of 100 students. | Example: Predicting average marks of all school students.   |


Question 2:What is sampling in statistics? Explain the differences between random
and stratified sampling.



**Sampling** is a statistical technique in which a small portion of individuals, called a **sample**, is selected from a large group known as the **population**, for the purpose of studying and drawing conclusions about the whole population.


Sampling is the process of selecting a subset of the population to obtain information and make estimates about the characteristics of the entire population.

Sampling is used in scientific research, business surveys, opinion polls, healthcare studies, education performance reviews, and market research.

## **Types of Sampling**

There are many sampling methods, but two important probability sampling methods are:

1. **Random Sampling**
2. **Stratified Sampling**

---

## **Random Sampling**

**Random sampling** is a technique where each member of the population has an **equal chance** of being selected in the sample. There is no bias in the selection process, and samples are chosen purely by chance.

 **Example of Random Sampling**

Suppose a school has 1,000 students, and we want to select 50 students randomly for a survey. We assign each student a number and use a lottery system or computer-based random number generator to pick 50 students. Every student has the same chance of selection.

 **Advantages**

* Simple and easy to use
* Free from personal bias
* Highly representative when population is homogeneous

 **Disadvantages**

* May not represent different groups within a population if the population is diverse

---

## **Stratified Sampling**

**Stratified sampling** divides the population into smaller groups called **strata** based on similar characteristics (e.g., age, gender, income level, region). Then **random samples** are taken from each group in proportion to their size.

 **Example of Stratified Sampling**

In the same school of 1,000 students:

* 600 boys
* 400 girls

If we want a sample of 50 students, we divide the population by gender (strata) and select students proportionally:

* From boys: (600/1000 × 50) = 30
* From girls: (400/1000 × 50) = 20

This ensures that both groups are fairly represented.

 **Advantages**

* Ensures representation of all groups
* More accurate and reliable than random sampling
* Useful when population is diverse

 **Disadvantages**

* More complex and time-consuming
* Requires detailed population information


 **Key Differences Between Random and Stratified Sampling**

| **Random Sampling**                                 | **Stratified Sampling**                                                    |
| --------------------------------------------------- | -------------------------------------------------------------------------- |
| Every member has an equal chance of being selected. | Population is divided into groups, and samples are chosen from each group. |
| Simple selection process.                           | More complex selection process.                                            |
| Best when population is homogeneous (similar).      | Best when population is heterogeneous (different groups).                  |
| May not represent all subgroups.                    | All important subgroups are represented proportionally.                    |
| Example: lottery selection of 50 students.          | Example: selecting sample based on gender proportion.                      |


Question 3:Define mean, median, and mode. Explain why these measures of central
tendency are important.


Measures of central tendency are statistical tools used to identify a single value that represents the center or average of a dataset. They help to summarize large amounts of data in a simple and understandable form. The three most commonly used measures of central tendency are **Mean**, **Median**, and **Mode**.

---

## **1. Mean**


 Mean is the arithmetic average of a set of values, obtained by adding all values and dividing the sum by the total number of values.

 **Formula**

[
\text{Mean} = \frac{\text{Sum of all observations}}{\text{Number of observations}}
]

 **Example**

If the marks of five students are: 70, 80, 65, 75, 90
Mean = (70 + 80 + 65 + 75 + 90) / 5 = 380 / 5 = **76**

---

## **2. Median**



Median is the middle value of an ordered dataset. It divides the data into two equal halves.

 **How to find median**

* Arrange the data in ascending order.
* If **n is odd**, median = middle value.
* If **n is even**, median = average of two middle values.

 **Example**

Data: 10, 15, 20, 25, 30
Ordered data: 10, 15, **20**, 25, 30
Median = **20** (middle value)

---

## **3. Mode**

Mode is the value that appears most frequently in a dataset.*

 **Example**

Data: 4, 8, 6, 8, 8, 10, 6
Mode = **8** (appears most often)

Mode is useful for identifying common or popular items (e.g., best-selling product size or most common color).

---

## **Importance of Mean, Median, and Mode**

 **1. Summarizes large data**

These measures reduce large sets of numbers into a single representative value for easier understanding and comparison.

 **2. Helps in decision making**

Used in business, economics, research, healthcare, and education for planning and forecasting.

 **3. Identifies typical or common behavior**

Mode helps identify the most frequent or popular choice (e.g., most common shoe size).

 **4. Useful for comparing different datasets**

Mean and median help compare performance or trends between groups (e.g., average salary across companies).

 **5. Helps understand data distribution**

Median is useful when data contains extreme values or outliers, where mean may be misleading.


Question 4: Explain skewness and kurtosis. What does a positive skew imply about the
data?


In statistics, understanding the shape and distribution of data is important for accurate interpretation and decision-making. Two important measures that describe the shape of a frequency distribution are **skewness** and **kurtosis**. These measures help us analyze how data values are spread around the central value (mean) and how extreme values behave.

 **Skewness**



Skewness is a measure of the asymmetry of the distribution of values in a dataset around its mean.*

If data is symmetric, the mean, median, and mode are equal. But if data is not symmetric, it is said to be **skewed**.

 **Types of Skewness**

1. **Positive Skew (Right-Skewed)**
2. **Negative Skew (Left-Skewed)**
3. **Zero Skew (Symmetrical distribution)**

 **Graphical Interpretation**

* **Positive Skew**: Tail on the right side is longer.
* **Negative Skew**: Tail on the left side is longer.

---

## **Kurtosis**


Kurtosis is a statistical measure that describes the shape of a distribution in terms of its peakedness or flatness compared to a normal distribution.*

It tells us how much of the data is concentrated in the center and how heavy or light the tails are.

### **Types of Kurtosis**

| Type            | Meaning                    | Shape                         |
| --------------- | -------------------------- | ----------------------------- |
| **Mesokurtic**  | Normal/average peak        | Bell-shaped                   |
| **Leptokurtic** | High peak, heavy tails     | More concentrated around mean |
| **Platykurtic** | Low flat peak, light tails | More spread out               |

---

## **What Does a Positive Skew Imply About the Data?**

A positive skew (right skew) means:

* The **tail of the distribution is longer on the right side**.
* A few extremely large values pull the tail to the right.
* **Mean > Median > Mode**
* Most data points are concentrated on the left side.

 **Example**

In income distribution:

* Many people have average or low income.
* A few people have very high salaries.
* This creates a right-skewed distribution (positive skew).

---

 **Importance of Skewness and Kurtosis**

* Helps understand real-world data patterns
* Identifies outliers and extreme values
* Supports decisions in finance, economics, business, psychology, etc.
* Helps choose the right statistical tests and models

Question 6: Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]


````python
import numpy as np

list_x = np.array([10, 20, 30, 40, 50], dtype=float)
list_y = np.array([15, 25, 35, 45, 60], dtype=float)

# Covariance matrix
cov_matrix = np.cov(list_x, list_y, ddof=1)
cov_xy = cov_matrix[0, 1]

# Correlation coefficient
corr_xy = np.corrcoef(list_x, list_y)[0, 1]

print("Covariance:", cov_xy)
print("Correlation Coefficient:", corr_xy)
````

Question 7: Write a Python script to draw a boxplot for the following numeric list and
identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]


````python
import matplotlib.pyplot as plt
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Calculate Q1, Q3 and IQR to find outliers
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = [x for x in data if x < lower_bound or x > upper_bound]

# Boxplot
plt.boxplot(data)
plt.title("Boxplot for Given Data")
plt.show()

print("Q1:", q1)
print("Q3:", q3)
print("IQR:", iqr)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)
````

Question 8: You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.
● Explain how you would use covariance and correlation to explore this
relationship.
● Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]


As a data analyst in an e-commerce company, one of the key tasks is to understand whether spending more on advertising leads to higher sales. Two statistical tools that help analyze the relationship between two numerical variables are **covariance** and **correlation**.

---

 **Using Covariance**

**Covariance** measures the **direction** of the relationship between two variables (i.e., whether they increase or decrease together).

* **Positive covariance** → when advertising spending increases, sales also increase.
* **Negative covariance** → when advertising increases, sales decrease.
* **Covariance near zero** → no relationship.

Covariance does **not** show the strength of the relationship, only its direction.

---

 **Using Correlation**

**Correlation coefficient** measures the **strength and direction** of the linear relationship between two variables.

* Value range: **–1 to +1**
* **+1** → perfect positive relationship
* **–1** → perfect negative relationship
* **0** → no relationship

Higher value closer to +1 indicates strong positive correlation (good for predicting sales using ad spend).

---

## **Python Code to Compute Correlation**

```python
import numpy as np

advertising_spend = np.array([200, 250, 300, 400, 500], dtype=float)
daily_sales = np.array([2200, 2450, 2750, 3200, 4000], dtype=float)

# Compute correlation coefficient
corr = np.corrcoef(advertising_spend, daily_sales)[0, 1]

print("Correlation Coefficient:", corr)
```



````python
import numpy as np

advertising_spend = np.array([200, 250, 300, 400, 500], dtype=float)
daily_sales = np.array([2200, 2450, 2750, 3200, 4000], dtype=float)

# Compute correlation coefficient
corr = np.corrcoef(advertising_spend, daily_sales)[0, 1]

print("Correlation Coefficient:", corr)
````

Question 9: Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.
● Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use.
● Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]


 **Understanding the Distribution of Customer Satisfaction Scores**

To analyze the survey data (scores from 1–10), we want to understand:

* The **central tendency** – where most scores lie
* The **spread / variability** – how far apart the scores are
* The **shape of the distribution** – skewed or normally distributed

 ** Statistics to Use**

| Statistic              | Purpose                                      |
| ---------------------- | -------------------------------------------- |
| **Mean (Average)**     | Shows the average satisfaction score         |
| **Median**             | Middle value; useful if outliers are present |
| **Mode**               | Most frequently occurring score              |
| **Standard Deviation** | Measures variability of satisfaction         |
| **Minimum & Maximum**  | Range of scores                              |
| **Count**              | Total number of survey responses             |

#### **Visualizations**

| Visualization            | Why Use it                                 |
| ------------------------ | ------------------------------------------ |
| Histogram            | Shows distribution and most common range   |
| Box Plot**             | Highlights median, quartiles, and outliers |
| Bar Chart (for mode) | Displays frequency of each rating          |

A histogram helps understand if most customers gave high ratings, low ratings, or mixed.

---

 **Python Code to Create a Histogram Using Matplotlib**


````python
import matplotlib.pyplot as plt

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Create histogram
plt.hist(survey_scores, bins=5)
plt.title("Distribution of Customer Satisfaction Scores")
plt.xlabel("Survey Score")
plt.ylabel("Frequency")
plt.show()
````
